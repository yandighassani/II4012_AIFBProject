# **Jenis Kekerasan**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('jenis_kekerasan.csv')

df = data.copy()

le = LabelEncoder()

df['jenis_kekerasan_encoded'] = le.fit_transform(df['jenis_kekerasan'])
df['jenis_kelamin_encoded'] = le.fit_transform(df['jenis_kelamin_korban'])

numeric_columns = ['jumlah_korban', 'jenis_kekerasan_encoded', 'jenis_kelamin_encoded']

correlation_matrix = df[numeric_columns].corr()

print(correlation_matrix)
```

```python
label = 'jumlah_korban'

correlation_with_label = correlation_matrix[label]

print(correlation_with_label)
```

```python
top_5_correlations = correlation_with_label.abs().sort_values(ascending=False).head(5)

print(top_5_correlations)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriks Korelasi')
plt.show()
```
